# Predict Canada's Per capita income in Year 2020

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Step 2: Load Dataset
df = pd.read_csv("canada_per_capita_income (1).csv")
```

```python
df.head(5)
```

```python
# Step 3: Define the features(X) and Target (y)
X = df[["year"]]
y = df[["per capita income (US$)"]]
```

```python
# Step 4: Train the model
reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
# Check model parameters
print("Slope (Coefficient):", reg.coef_)
print("Intercept:", reg.intercept_)
```

```python
# Make prediction
pred = reg.predict([[2020]])
pred
```

```python
# step visualization

plt.scatter(X, y, color='blue', label="Actual data")
plt.plot(X, reg.predict(X), color='red', label="Regression Line")
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.legend()
plt.show()
```
