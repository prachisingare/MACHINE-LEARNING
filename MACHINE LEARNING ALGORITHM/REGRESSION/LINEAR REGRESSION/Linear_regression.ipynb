```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\Prachi\Downloads\homeprices.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price(US$)')
plt.scatter(df[['area']] , df[['price']], color = 'red', marker = '+')
plt.show()
```

```python
# craete linear regression model
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3000]])
```
